### Chill Factor

$$WCI = (10\sqrt v - v + 10.5) * (33 - T _m)$$

```python
data.head()
```

```python
from math import sqrt
```

```python
def wci(x): # x is a row of data
    velocity = x.WindGustSpeed
    return (10 * sqrt(velocity) - velocity + 10.5) * (33 - x.MinTemp)
```

```python
data['WCI'] = data.apply(wci, axis=1)
```

```python
data
```

```python
data_groupedByMonth = data.groupby(by=["Month"]).mean(numeric_only=True)

data_groupedByMonth.head()
```

```python
data_groupedByMonth.sort_values("WCI", ascending=False)
```
