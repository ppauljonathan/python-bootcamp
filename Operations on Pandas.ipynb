## Lambda Functions using apply

```python
data.Rainfall
```

```python
data.Rainfall.apply(lambda x: 'Rainy' if x > 50 else 'Not Rainy')
```

```python
data['IsRainyDay'] = data.Rainfall.apply(lambda x: 'Rainy' if x > 50 else 'Not Rainy')
```
